```python
import psycopg2 as db
```

```python
conn_string = "dbname='nifi_db' host='localhost' user='root' password='root'"

conn=db.connect(conn_string)

cursor=conn.cursor()
```

```python
query = "insert into users(id,name,street,city,zip_code) values ({}, '{}', '{}', '{}', '{}')".format(1, 'Big bird', 'Sesame Street', 'Fakeville', '12345')
```

```python
cursor.mogrify(query)
```

```python
cursor.execute(query)
conn.commit()
```

```python
from faker import Faker
```

```python
fake=Faker()
data=[]
```

```python
for i in range(1, 1000):
    data.append((i, fake.name(), fake.street_address(), fake.city(), fake.zipcode()))
    
```

```python
data
```

```python
data_for_db=tuple(data)
```

```python
query = "insert into users(id,name,street,city,zip_code) values (%s, %s, %s, %s, %s)"

cursor.mogrify(query, data_for_db[1])
```

```python
cursor.executemany(query, data_for_db)
conn.commit()
```
