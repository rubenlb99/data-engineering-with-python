```python
from faker import Faker
output=open('data.csv', mode='w')
fake=Faker()
header=['name','age','street','city','state','zip','lng','lat']
my_writer=csv.writer(output)
my_writer.writerow(header)
for r in range(1000):
    my_writer.writerow([fake.name(), fake.random_int(min=18, max=80, step=1), fake.street_address(),fake.city(), fake.state(), fake.zipcode(), fake.longitude(), fake.latitude()])
    
output.close()
```

```python
with open('data.csv') as f:
    myreader=csv.DictReader(f)
    headers=next(myreader)
    for row in myreader:
        print(f'{row['name']}, {row['age']}')
```
