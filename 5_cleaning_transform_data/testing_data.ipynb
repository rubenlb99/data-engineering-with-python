<h1>Reading and understanding data</h1>

```python
import pandas as pd
```

```python
df = pd.read_csv("scooter.csv")

df.head(10)
```

```python
#get sum of nulls per column
df.isnull().sum()
```

```python
#intervals
df['trip_id'].value_counts(bins=10)
```

```python
#delete region id as its the same value in whole df
df.drop(columns=['region_id'], inplace=True)
```

```python
#get rows where a specific column is null
df[df['start_location_name'].isnull()]
```

```python
#drop values where a specific column is na
df.dropna(subset=['start_location_name'], inplace=True)
```

```python
#get subset where end location name is null
stop = df[df['end_location_name'].isnull()]

#value to substitute the rows
value = {'end_location_name':'Stop St.'}

#fill the null values in subset with value
stop.fillna(value=value)

```

<br><br>
<h1>Creating and modifying columns</h1>

```python
#make all column names lowercase
df.columns=[x.lower() for x in df.columns] 
print(df.columns)
```

```python
#make all months capital
df['month'] = df['month'].str.upper()

df['month'].head()
```

```python
#change dates to datetime type
df['started_at'] = pd.to_datetime(df['started_at'], format='%m/%d/%Y %H:%M')
df.dtypes
```

<br><br>
<h1>Enriching Data</h1>

```python
most_frecuent_starting_locations = pd.DataFrame(df['start_location_name'].value_counts().head())
most_frecuent_starting_locations.reset_index(inplace=True)
most_frecuent_starting_locations.columns = ['address', 'count']

most_frecuent_starting_locations
```

```python
#adding new column only with the street, replace "@" with "and"
n=most_frecuent_starting_locations['address'].str.split(pat=',',n=1, expand=True)
replaced=n[0].str.replace("@", "and")
most_frecuent_starting_locations['street']=replaced
most_frecuent_starting_locations
```

```python
geo = pd.read_csv('geocodedstreet.csv')
geo
```

```python
merged = pd.merge(most_frecuent_starting_locations, geo, on = 'street')
merged
```
